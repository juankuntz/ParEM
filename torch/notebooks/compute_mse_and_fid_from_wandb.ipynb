```python
import wandb
import numpy as np
import sys
import torch
sys.path.append("../")
import parem
from parem.utils import get_mnist
from pathlib import Path
import parem.stats as stats
from collections import defaultdict
api = wandb.Api()
%load_ext autoreload
%autoreload 2
```

```python
run_names = {
    'pga_smallbatch' : ['jenninglim/PGA_dev_MNIST_10000/1iqa38rj'],
    'abp' : ['jenninglim/PGA_dev_MNIST_10000/fg1jiyre'],
    'shortrun' : ['jenninglim/PGA_dev_MNIST_10000/232efswu'],
    'vae' : ['jenninglim/PGA_dev_MNIST_10000/d671fjw4'],
}
mse_dict = defaultdict(list)
fid_dict = defaultdict(lambda : defaultdict(list))
n_missing_img = 10
N_IMAGES = 300
```

```python
for k, names in run_names.items():
    for name in names:
        run = api.run(name)
        best_i = run.history()['loss'].argmin()
        run.file(f"{best_i}.cpt").download(replace=True)
        alg = torch.load(f"{best_i}.cpt")
        DATASET_PATH = Path('./datasets/MNIST')
        
        dataset = get_mnist(DATASET_PATH, N_IMAGES, train=False)  # Load dataset

        
        images = dataset[:n_missing_img][0]
        mask = torch.ones(dataset.height, dataset.width, dtype=torch.bool)

        for i in range(10, 22):
          for j in range(10, 22):
                mask[i, j] = False

        reconstructed_image = alg.reconstruct(images, mask)
        mse = ((images - reconstructed_image) ** 2).mean([-1, -2, -3]).mean(0).item()
        print(f"mse achieved {mse}")
        mse_dict[k].append(mse)

        n_componentss = [1 + 100 * i for i in range(10)]

        idx = np.random.choice(len(dataset), size=500)
        for n_components in n_componentss:
            model_samples = alg.synthesize_images(N_IMAGES, show=True, approx_type='gmm', n_components=n_components)
            data_samples = torch.stack([dataset[id][0] for id in idx], dim=0)
            gmm_fid = stats.compute_fid(data_samples,
                                        model_samples,
                                        n=N_IMAGES,
                                        nn_feature=None,
                                        device='cpu')
            print(n_components, gmm_fid)
            fid_dict[k][f'{n_components}'].append(gmm_fid)
```
